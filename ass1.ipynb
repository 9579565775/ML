```python
def remove_outlier(data1,col):
    Q1=data1[col].quartile(0.25)
    Q3=data1[col].quartile(0.75)
    IQR=Q3-Q1
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR
    
    data1[col]=np.where(data1[col]<lower_bound,lower_bound.data1[col])
    data1[col]=np.where(data1[col]<upper_bound,upper_bound.data1[col])
    return data1
```

```python
def treat_outliers(data):

    col_list = data.select_dtypes(include=[np.number]).columns
    for col in col_list:
        data = remove_outliers(data, col)
    return data
```

```python
data1=treat_outliers(data)
```
