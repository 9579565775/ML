```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('sales.csv',encoding='latin1')
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.drop(columns=['ADDRESSLINE1','ADDRESSLINE2','CITY','CUSTOMERNAME','STATE','POSTALCODE','PHONE','TERRITORY','CONTACTLASTNAME','CONTACTFIRSTNAME','STATUS','STATE','ORDERDATE'],axis=1,inplace=True)
```

```python
data.head()
```

```python
data.shape
```

```python
dealsize=pd.get_dummies(data['DEALSIZE'])
```

```python
dealsize
```

```python
country=pd.get_dummies(data['COUNTRY'])
```

```python
country
```

```python
productline=pd.get_dummies(data['PRODUCTLINE'])
```

```python
productline
```

```python
data.shape
```

```python
data=pd.concat([data,dealsize,country,productline],axis=1)
```

```python
data.drop(['DEALSIZE','PRODUCTLINE','COUNTRY'],axis=1,inplace=True)
```

```python
data.shape
```

```python
data['PRODUCTCODE']=pd.Categorical(data['PRODUCTCODE']).codes
```

```python
data.describe()
```

```python
data.describe().T
```

```python
data.head()
```

```python
from sklearn.cluster import KMeans
```

```python
disortion=[]
k=range(1,10)
for n in k:
    km=KMeans(n_clusters=n)
    km.fit(data)
    disortion.append(km.inertia_)
```

```python
plt.plot(k,disortion,'-bx')
plt.xlabel('k')
plt.ylabel('Disortion')
plt.title('Elbow Curve')
plt.show()
```

```python
plt.plot(k,disortion,'-bx')
plt.xlabel('k')
plt.ylabel('Disortion')
plt.title('Elbow Curve')
plt.show()
```

```python
x_train=data.values
model=KMeans(n_clusters=4,random_state=2)
model.fit(x_train)
pred=model.predict(x_train)
```
